```python
class Point:
    """Represents a point in 2-D space."""
    
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def __str__(self):
        return f'Point({self.x}, {self.y})'
    
    def __eq__(self, other):
        return (self.x == other.x) and (self.y == other.y)

    def translate(self, dx, dy):
        self.x += dx
        self.y += dy
    
    def translated(self, dx=0, dy=0):
        point = copy(self)
        point.translate(dx, dy)
        return point
```

```python
class Line:
    def __init__(self, p1, p2):
        self.p1 = p1
        self.p2 = p2
        
    def __str__(self):
        return f'Line({self.p1}, {self.p2})'
    
    # Exercise 1:

    def __eq__(self, other):
       return ((self.p1 == other.p1) and (self.p2 == other.p2)) or ((self.p1 == other.p2) and (self.p2 == other.p1))
    
    # Exercise2:

    def midpoint(self):
        point1 = (self.p1.x + self.p2.x) // 2
        point2 = (self.p1.y +self.p2.y) // 2
        return Point(point1, point2)

    def draw(self):
        jumpto(self.p1.x, self.p1.y)
        moveto(self.p2.x, self.p2.y)
```

```python
class Rectangle:
    """Represents a rectangle. 

    attributes: width, height, corner.
    """
    def __init__(self, width, height, corner):
        self.width = width
        self.height = height
        self.corner = corner
        
    def __str__(self):
        return f'Rectangle({self.width}, {self.height}, {self.corner})'
    

    def make_points(self):
        p1 = self.corner
        p2 = p1.translated(self.width, 0)
        p3 = p2.translated(0, self.height)
        p4 = p3.translated(-self.width, 0)
        return p1, p2, p3, p4
    
    def make_lines(self):
        p1, p2, p3, p4 = self.make_points()
        return Line(p1, p2), Line(p2, p3), Line(p3, p4), Line(p4, p1)
    
    def draw(self):
       lines = self.make_lines()
       for line in lines:
         line.draw()

    def grow(self, dwidth, dheight):
        self.width += dwidth
        self.height += dheight
        
    def translate(self, dx, dy):
        self.corner.translate(dx, dy)

    # Exercise 3:
    
    def midpoint(self):
        x1 = self.corner.x
        y1 = self.corner.y

        x = x1 + (self.width) // 2
        y = y1 + (self.height) // 2
        return Point(x, y)
     # Exercise 4:

    def make_cross(self):
        
        lines = self.make_lines()

        mid1 = lines[0].midpoint()
        mid2 = lines[1].midpoint()
        mid3 = lines[2].midpoint()
        mid4 = lines[3].midpoint()

        line1 = Line(mid1, mid3)
        line2 = Line(mid2, mid4)

        return[line1, line2]
        

  

    
       
       
```

### Exercise 5

Write a definition for a class named `Circle` with attributes `center` and `radius`, where `center` is a Point object and `radius` is a number.
Include special methods `__init__` and a `__str__`, and a method called `draw` that uses `jupyturtle` functions to draw the circle.

```python
import math
```

```python
class Circle:

    def __init__(self, center, radius):
        self.center = center
        self.radius = radius

    def __str__(self):
        return f"Circle, center = {self.center}, radius = {self.radius}"
    
    def draw(self):
        jumpto(self.center.x + self.radius, self.center.y)
        left(90)
        step = (math.pi * 2 * self.radius) / 360
        for _ in range(0, 360):
            forward(step)
            left(1)
```

You can use the following function, which is a version of the `circle` function we wrote in Chapter 4.

```python
from jupyturtle import make_turtle, forward, left, right
import math

def draw_circle(radius):
    circumference = 2 * math.pi * radius
    n = 30
    length = circumference / n
    angle = 360 / n
    left(angle / 2)
    for i in range(n):
        forward(length)
        left(angle)
```
