## 1. توليد كلمة مرور

اكتب إجراءً يقبل المعاملات التالية ليُنتِج **كلمة مرور** (Password):

- الطول
- الأحرف الصغيرة
- الأحرف الكبيرة
- الأرقام
- الحد الأدنى من الأرقام
- الرموز: `!@#$%^&*`
- الحد الأدنى من الرموز

```python
def generate_password(length: int, ...) -> str:
    ...
```

````python
      # اكتب

def generate_password(length, lower_case, upper_case, nums, min_nums, symbols, min_sym ):
    random.seed(47)

    if min_nums + min_sym < length:

      pass_nums = random.choices(population = nums, k = min_nums)
      pass_symbols = random.choices(population = symbols, k = min_sym)
      pass_letters = random.choices(population = lower_case + upper_case + nums + symbols, k = (length - (len(pass_symbols) + len(pass_nums))))
      all_pass =  pass_nums + pass_symbols + pass_letters
      random.shuffle(all_pass)
      password = ''.join(all_pass)

      return password
    else:
       return f"minimum numbers is longer than password length!"

````

````python
generate_password(10,'abcdefg','ABCDEFG','123456789',4,'!@#$%^&*()',2)
````

## 2. توليد عبارة مرور

اكتب إجراءً آخر يقبل المعاملات التالية ليُنتِج [**عبارة مرور** (Passphrase)](https://proton.me/blog/what-is-passphrase)، بحيث تكون سهلة الحفظ على البشر:

- عدد الكلمات
- فاصل الكلمات
- تكبير الحرف الأول من كل كلمة
- تضمين رقم

```python
def generate_passphrase(...):
    ...
```

````python
from faker import Faker
def generate_passphrase(number_of_words, spaces, capitalize_first, include_num):
    
    random.seed(47)

    fake = Faker()
    words = ['']
    for _ in range(100):
       words.append(fake.word())

    result1 = random.choices(population = words, k = number_of_words)

    if capitalize_first:
        
            result1 = [i.capitalize() for i in result1]

    if include_num:
        result2 = str(random.randint(0, 10))
        result1.append(result2)

    result = spaces.join( result1 )

    
    return result    # اكتب هنا
````

````python
generate_passphrase(10, '_', True, True)
````

````python
generate_passphrase(7, '!', False, True)
````

````python
generate_passphrase(8, ':', False, False)
````
